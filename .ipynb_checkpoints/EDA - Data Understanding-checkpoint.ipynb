## Exploratory Data Analysis - Data Understanding

The aim of this exercise is to analyze a dataset that contains previous observation of an ecommerce shipping service. Understand behaviour and find patterns between the different attributes.

Originally the dataset consists of 10,999 records and 12 different attributes.

#### Dataset Attributes:

- ID: ID Number of Customers.
- Warehouse block: The Company have big Warehouse which is divided in to block such as A,B,C,D,E.
- Mode of shipment:The Company Ships the products in multiple way such as Ship, Flight and Road.
- Customer care calls: The number of calls made from enquiry for enquiry of the shipment.
- Customer rating: The company has rated from every customer. 1 is the lowest (Worst), 5 is the highest (Best).
- Cost of the product: Cost of the Product in US Dollars.
- Prior purchases: The Number of Prior Purchase.
- Product importance: The company has categorized the product in the various parameter such as low, medium, high.
- Gender: Male and Female.
- Discount offered: Discount offered on that specific product.
- Weight in gms: It is the weight in grams.
- Reached on time: It is the target variable, where 1 Indicates that the product has NOT reached on time and 0 indicates it has - reached on time.

In later stages, different Machine Learning models will be implemented with the aim of finding the one that provides the most accurate predictions.



```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv('./train.csv')
df.head(4)
```

```python
def checkNullValues():
    n_entries = df.shape[0]
    
    for column in df:
        print('{} | {} | {}'.format(df[column].name, 
                            len(df[df[column].isnull()]) / (1.0*n_entries),
                            df[column].dtype))
              
checkNullValues()
```

```python
def checkCategoricalValues(categoricalColumns):
    for column in categoricalColumns:
        print('{} | Classes: {}'.format(df[column].name, df[column].unique()))

categorical_columns = ['Gender', 'Warehouse_block', 'Mode_of_Shipment', 'Product_importance', 'Customer_rating']
numerical_columns = ['Customer_care_calls', 'Cost_of_the_Product', 'Prior_purchases',\
                    'Discount_offered', 'Weight_in_gms']

target = 'Reached.on.Time_Y.N'

checkCategoricalValues(categorical_columns)
```

```python
#Proportions of categorical variables according to target variable
def proportionsTargetVariable(column):
    
    values = df[column].value_counts()
    keys = df[column].value_counts().keys()
    
    for i in range(values.size): 
        
        onTime = df[(df[column] == keys[i]) & (df['Reached.on.Time_Y.N'] == 1)].shape[0]
        delayed = df[(df[column] == keys[i]) & (df['Reached.on.Time_Y.N'] == 0)].shape[0]
        
        print('{} | On Time: {} | Delayed: {}'.format(keys[i], onTime / (1.0 * values[i]), delayed / (1.0 * values[i]) ))
        print('Total: {}'.format(onTime + delayed))

proportionsTargetVariable('Warehouse_block')
```

# Conclusions:

### About the dataset:

- There are no null values for any of the attributes.
- The entries represent a single purchase from a 10,999 customer database, although customers might have made prior purchases.
- The only attribute that seems to have a normal distribution is the number of customer calls.
- Some attributes like prior purchases and discount offered seem to have outliers.
- 60% of shipments were delivered on time and 40% were delayed.

### Attributes:

##### Categorical Attributes:
- When considering the target variable every categorical variable seems to follow the distribution of the target variable, that is to say that every category is distributed 60% delivered on time and 40% delayed.
- 33% of shipments are handled by the Warehouse block F and the rest are distributed equally between the other blocks.
- Gender is equally distributed between males and females.
- 68% of shipments were made by ships and the rest was distributed between flights and road trips.
- Most packages were of low or medium importance, whereas only 9% of them were highly important.

#### Numerical Attributes:
- All customers in the dataset have made at least 2 customer care calls.
- The price of the products oscile between around 50USD to 300USD. With no outliers.
- All customers had made at least 2 prior purchases. For this attribute there seems to be a few outliers as the maximum of the boxplot is 5 prior purchases. More purchases might be considered.
- Discounts offered seems to be normally distributed between values 0 to 20%. However 20% of the entries have a discount over the 20%.
- There seems to be a relationship between the number of customer care calls and the cost of the product. The more expensive the greater the number of customer calls.

#### Numerical Distributions according to Target Variable:
- It seems that those customer whose packages were delayed were more likely to make more customer care calls.
- Products with a discount between 0-13% were more likely to be delayed.
- The heavier the products the more likely they were to be delayed.
- There is a relationship betwwen the number of customer care calls and the cost of the product.
